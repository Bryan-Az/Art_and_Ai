# Geography and Demography within the NGA Data

### Top Nations Present in National Gallery of Art in Washington, DC (EDA)

```python
# I need to import a copy of the constituents 'nationality' column from the non_latin_art.csv dataset recently added into 
latinamerican_nationalities = pd.read_csv('../../data_samples/LaArt/constituents_nationalities.csv')
non_latinamerican_nationalities = pd.read_csv('../../data_samples/nonLaArt/constituents_nationalities.csv')
```

```python
la_artist_origin = latinamerican_nationalities.value_counts(normalize=True)
nonla_artist_origin = non_latinamerican_nationalities.value_counts(normalize=True)
```

```python
la_artist_origin[:10].plot(kind='bar')
```

The above plot shows the top 10 latin-american nationalities of artists within the NGA's dataset, out of all latin-american data within the dataset. (~350 latinamerican data points)

```python
nonla_artist_origin[:10].plot(kind='bar')
```

The above plot shows the top 10 non latin-american nationalities of artists within the NGA's dataset, out of all non latin-american data within the dataset. (~37,200 non latin-american data points)
